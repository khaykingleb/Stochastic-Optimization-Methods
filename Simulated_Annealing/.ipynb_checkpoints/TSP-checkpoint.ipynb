```python
def distance(x, n): 
    d = dict()
    
    for i in range(n):
        city = dict()
        for j in range(n): # для измерения расстояния используем евклидовую метрику
            if i == j:
                continue
            c_a = x[i][1]
            c_b = x[j][1]
            dist = np.sqrt((c_a[0] - c_b[0]) ** 2 + (c_a[1] - c_b[1]) ** 2)
            city[x[j][0]] = dist
        d[x[i][0]] = city
    
    return d
```

```python
def F(path, cities): # высчитываем общее расстояние путешествия
    dist = 0
    
    for i in range(len(path) - 1):
        dist += cities[path[i]][path[i + 1]]
    
    dist += cities[path[i + 1]][path[0]]
    
    return dist
```

```python
def SA(path, T):
    path_hat = path
    n = len(path_hat)
    np.random.seed(33412)
    np.random.shuffle(path_hat)
    T_0 = T
    k = 1
    
    for i in range(100000):
        path_tilda = G(path_hat, n)
        delta = F(path_tilda, cities_d) - F(path_hat, cities_d)
        prob = np.exp(- delta / T)
        
        if (delta < 0) or (prob >= np.random.random()):
            path_hat = path_tilda
        
        T = T_0 / np.log(k + 1)# 
        k += 1    
        
    return path_hat
```

```python
def G(path, n):
    pos = path.copy()
    while True:
        i = np.random.randint(0, n - 1)
        j = np.random.randint(0, n - 1)
    
        if i != j:
            break
            
    pos[i], pos[j] = pos[j], pos[i]
    return pos
```

```python
path_opt = SA(names, 100)
```
